```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("iris-data.csv")
```

```python
data.dropna(subset=["petal_width_cm"],inplace=True)
data.isnull().sum()
```

```python
data["class"].value_counts()
```

```python
data['class'].replace(["Iris-setossa","versicolor"],["Iris-setosa","Iris-versicolor"],inplace=True)
data["class"].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data["class"],palette="muted")
plt.plot()
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(data,height=2.5,hue="class")
plt.plot()
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(data.corr(),cmap="coolwarm",annot=True)
plt.yticks(rotation=0)
plt.xticks(rotation=30)
plt.plot()
```

```python
sns.set(style="white", palette="muted", color_codes=True)
fig = plt.figure(figsize=(10,15))
sns.jointplot(data["sepal_length_cm"],data["sepal_width_cm"],kind="kde",color="purple")
plt.plot()
```

```python
plt.figure(figsize=(10,15))
sns.jointplot(data["petal_length_cm"],data["petal_width_cm"],kind="kde",color="purple")
plt.plot()
```

```python
fig = plt.figure(figsize=(20,10))

axes = fig.subplots(2,2)

sns.distplot(data["sepal_length_cm"], color="purple",ax=axes[0][0])
sns.distplot(data["sepal_width_cm"], color="teal",ax=axes[0][1])

sns.distplot(data["petal_length_cm"], color="red",ax=axes[1][0])
sns.distplot(data["petal_width_cm"], color="blue",ax=axes[1][1])

plt.plot()
```

```python
X = data.drop(["class"],axis=1)
y = data["class"]

print(X.head())
print("########################")
print(y.head())
```

```python
xtr,xts,ytr,yts = train_test_split(X,y,test_size=0.3)
```

```python
print(xtr.shape)
print(ytr.shape)
print(xts.shape)
print(yts.shape)
```

```python
encode = LabelEncoder()
encode.fit_transform(ytr)
encode.fit_transform(yts)
```

```python
## Logistic Regression 

lr_model = LogisticRegression()
lr_model.fit(xtr,ytr)
lr_predict = lr_model.predict(xts)
print('Logistic Regression - ',accuracy_score(lr_predict,yts))
```

```python
## SVM Model
from sklearn.svm import SVC

svm_model = SVC(kernel='linear')
svm_model.fit(xtr,ytr)
svc_predict = svm_model.predict(xts)
print('SVM - ',accuracy_score(svc_predict,yts))
```

```python
## Naive Bayes
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(xtr,ytr)
nb_predict = nb_model.predict(xts)
print('Naive bayes - ',accuracy_score(nb_predict,yts))
```

```python

## Descision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(max_leaf_nodes=3)
dt_model.fit(xtr,ytr)
dt_predict = dt_model.predict(xts)
print('Decision Tree - ',accuracy_score(dt_predict,yts))
```
