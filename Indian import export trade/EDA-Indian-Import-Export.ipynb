```python
def create_summary(dataframe):
    print("Shape of dataframe",dataframe.shape)
    
    summary = pd.DataFrame()
    summary['dtypes'] = dataframe.dtypes
    summary['Unique_Values'] = dataframe.nunique().values
    summary['Missing'] = dataframe.isnull().sum().values
    summary['First Value'] = dataframe.loc[0].values
    summary['Second Value'] = dataframe.loc[1].values
    summary['Third Value'] = dataframe.loc[2].values
    
    return summary
```

```python
create_summary(importdata)
```
